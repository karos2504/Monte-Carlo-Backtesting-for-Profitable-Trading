```python
import numpy as np
import matplotlib.pyplot as plt

# Set the shape parameters for the Pareto distributions
shape = 1.0

# Generate numbers values for Pareto distributions
size = 10000
pareto_numbers = np.random.pareto(shape, size=size)

# Plot the Pareto distribution
plt.figure(figsize=(10, 6))
plt.hist(pareto_numbers, bins=50, density=True, alpha=0.6, color='g')
plt.title(f'Pareto Distribution (shape={shape})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.yscale('log')  # Use logarithmic scale for y-axis
plt.grid()
plt.show()
```

```python
# Parameters for synthetic price series
shape = 1.0     # Shape parameter (alpha), controls tail heaviness
scale = 0.01    # Scale parameter, controls spread of returns
size = 1000      # Number of trading days
start_price = 100  # Starting price of the asset

# Simulate log-returns using Pareto distribution
log_return_pareto = (np.random.pareto(shape, size=size) + 10) * scale  # Shift and scale

directions = np.random.choice([-1, 1], size=size)  # Randomly assign positive or negative direction
log_return_pareto *= directions

prices_pareto = start_price * np.exp(np.cumsum(log_return_pareto))

# Plot the synthetic price series
plt.figure(figsize=(10, 6))
plt.plot(prices_pareto, label=f'Pareto Distribution Returns (shape={shape}, scale={scale})', color='blue')
plt.title("Synthetic Price Series using Pareto Distribution")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```
