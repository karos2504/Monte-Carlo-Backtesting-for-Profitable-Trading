```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the log-returns distribution
mu = 0      # Mean of the underlying normal distribution
sigma = 0.1 # Standard deviation of the underlying normal distribution
size = 1000 # Number of random samples to generate
start_price = 100  # Starting price of the asset

# Simulate log-returns
log_returns = np.random.normal(loc=mu, scale=sigma, size=size)
prices = start_price * np.exp(np.cumsum(log_returns))

# Plot the simulated price series
plt.figure(figsize=(10, 6))
plt.plot(prices, label=f'Simulated Price (mu={mu}, sigma={sigma})', color='blue')
plt.title('Simulated Asset Price Series using Log-Return Distribution')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
