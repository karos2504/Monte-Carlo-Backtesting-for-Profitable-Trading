# Triangular Distribution

The triangular distribution is defined by a minimum, maximum, and mode — useful for bounded return scenarios.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate numbers based on Triangular distribution
low = 0
high = 1
mode = 0.5
size = 100000

triangular_numbers = np.random.triangular(low, mode, high, size=size)

# Plot histogram of triangular distribution
plt.figure(figsize=(10, 6))
plt.hist(triangular_numbers, bins=100, density=True, alpha=0.6, color='blue')
plt.title('Histogram of Triangular Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

## 1 — Generating Triangular Samples & Plotting

```python
# Parameters for synthetic price series
low = -0.01  # Minimum daily return (-1%)
high = 0.01  # Maximum daily return (1%)
mode = 0.001  # Most likely daily return (0.1%)
size = 1000      # Number of trading days
start_price = 100  # Starting price of the asset

# Simulate log-returns using Triangular distribution
triangular_returns = np.random.triangular(low, mode, high, size=size)
prices_triangular = start_price * np.exp(np.cumsum(triangular_returns))

# Plot the synthetic price series
plt.figure(figsize=(10, 6))
plt.plot(prices_triangular, label=f'Triangular Distribution Returns (low={low}, mode={mode}, high={high})', color='blue')
plt.title("Synthetic Price Series using Triangular Distribution")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```
