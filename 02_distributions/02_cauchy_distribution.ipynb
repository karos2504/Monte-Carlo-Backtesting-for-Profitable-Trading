# Cauchy Distribution

The Cauchy distribution has heavier tails than the normal, making it useful for modeling extreme price movements.

```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.standard_cauchy(size=100000)*0.02
random_numbers = np.clip(random_numbers, -10, 10)

plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=100, density=True, alpha=0.6, color='g')
plt.title('Histogram of Clipped Cauchy Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.yscale('log')
plt.grid(True)
plt.show()
```

## 1 â€” Generating Cauchy Samples & Plotting

```python
# Parameters for synthetic price series
loc = 0        # Mean of log-returns
scale = 0.02     # Standard deviation of log-returns
size = 1000      # Number of trading days
start_price = 100  # Starting price of the asset

# Simulate log-returns using Cauchy distribution
cauchy_returns = np.random.standard_cauchy(size=size)*scale + loc
log_cauchy_returns = np.clip(cauchy_returns, -0.1, 0.1)
prices_cauchy = start_price * np.exp(np.cumsum(log_cauchy_returns))

# Plot the synthetic price series
plt.figure(figsize=(10, 6))
plt.plot(prices_cauchy, label=f'Cauchy Distribution Returns (loc={loc}, scale={scale})', color='blue')
plt.title("Synthetic Price Series using Cauchy Distribution")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```
