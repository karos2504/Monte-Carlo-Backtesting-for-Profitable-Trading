# Laplace Distribution

The Laplace (double exponential) distribution has sharper peaks and heavier tails than the Gaussian, better capturing financial return distributions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values with Laplace distribution
loc, scale = 0, 1  # location and scale parameters
laplace_values = np.random.laplace(loc, scale, 1000)

# Plot histogram of Laplace distribution
plt.hist(laplace_values, bins=50, density=True, edgecolor='black', alpha=0.7)
plt.title("Laplace Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

## 1 — Generating Laplace Samples & Plotting

```python
# Generate values for distribution
loc, scale = 0, 1  # location and scale parameters

laplace_values = np.random.laplace(loc, scale, 1000)

normal_values = np.random.normal(loc, scale, 1000)

# Plot histogram of Laplace distribution
plt.hist(laplace_values, bins=100, density=True, edgecolor='black', alpha=0.5, label='Laplace')
plt.hist(normal_values, bins=100, density=True, edgecolor='black', alpha=0.5, label='Normal')
plt.title("Laplace vs Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## 2 — Laplace vs Normal: Side-by-Side Comparison

```python
# Parameters for synthetic price series
loc = 0        # Mean of log-returns
scale = 0.02     # Standard deviation of log-returns
size = 1000      # Number of trading days
start_price = 100  # Starting price of the asset

# Simulate log-returns using Laplace distribution
laplace_returns = np.random.laplace(loc=loc, scale=scale, size=size)
prices_laplace = start_price * np.exp(np.cumsum(laplace_returns))

# Plot the synthetic price series
plt.figure(figsize=(10, 6))
plt.plot(prices_laplace, label=f'Laplace Distribution Returns (loc={loc}, scale={scale})', color='blue')
plt.title("Synthetic Price Series using Laplace Distribution")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## 3 — Synthetic Price Series Using Laplace Returns

```python
# Parameters for synthetic price series
loc = 0.001        # Mean of log-returns
scale = 0.02     # Standard deviation of log-returns
size = 1000      # Number of trading days
start_price = 100  # Starting price of the asset

# Simulate log-returns using Laplace distribution
laplace_returns = np.random.laplace(loc=loc, scale=scale, size=size)
prices_laplace = start_price * np.exp(np.cumsum(laplace_returns))

# Simulate log-returns using Normal distribution
normal_log_returns = np.random.normal(loc=loc, scale=scale, size=size)
prices_lognormal = start_price * np.exp(np.cumsum(normal_log_returns))

# Plot the synthetic price series
plt.figure(figsize=(10, 6))
plt.plot(prices_laplace, label=f'Laplace Distribution Returns (loc={loc}, scale={scale})', color='blue')
plt.plot(prices_lognormal, label=f'Normal Distribution Returns (mean={loc}, sigma={scale})', color='red')
plt.title("Synthetic Price Series using Laplace and Normal Distributions")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```
